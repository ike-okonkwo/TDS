```python
import pandas as pd
import math
```

```python
un_data =  pd.read_csv('data/un.csv')
gdp_data = pd.read_csv('data/ny.gdp.mktp.cd_Indicator_en_csv_v2.csv', skiprows = [0])
```

To be able to do this, you actually need to reset the index of both Data sets to be the country. This will make things much easier. After reindexing the dataframes, you can then use a for loop to assign values in the right place.


```python
## UN Data
del un_data['#']
un_data.set_index('Country', inplace = True)
un_data['Gap ann_gdp'] = 0
un_data.head(10)
```

```python
# GDP Data
gdp_data.fillna(value=0, inplace=True)
gdp_data.set_index('Country Name', inplace = True)
del gdp_data['Country Code']
del gdp_data['Indicator Name']
del gdp_data['Indicator Code']
gdp_data.head()
```

```python
for country in un_data.index:
    year = un_data.ix[country]['Year']
    if country in gdp_data.index:
        un_data.ix[country]['Gap ann_gdp'] = gdp_data.ix[country][str(year)]
```

```python
## lets look at the results 
un_data.head(10)
```

Notes : If notice, there are some countries in un_data that are not in gdp_data. Those countries will have 0 for the 'Gap ann_gdp'. An example is Anguilla . You can also do this using SQl-like joins, but you may have to write some complex joins

```python
un_data[un_data['Gap ann_gdp'] == 0]
```

```python
def log_transform(dt):
    if dt == 0:
        return 0
    else:
        return math.log10(dt)
```

```python
un_data['Gap ann_gdp_log'] = un_data['Gap ann_gdp'].apply(log_transform)
```
